```python

import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer


```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text = "Natural Language Processing is transforming how computers understand human language!"
```

```python
# Step 1: Lowercase
text = text.lower()
print("Lowercased:", text)
```

```python
# Step 2: Tokenize into words
tokens = word_tokenize(text)
print("tokens:", tokens)
```

```python
# Step 2: Tokenize into words
tokens = word_tokenize(text)
print("tokens:", tokens)
```

```python
# Step 3: Remove punctuation
tokens = [t for t in tokens if t not in string.punctuation]
print("No punctuation:", tokens)
```

```python
# Step 4: Remove stopwords (the, is , how, ...)
stop_words = set(stopwords.words('english'))
tokens = [t for t in tokens if t not in stop_words]
print("No stopwords:", tokens)
```

```python
# Step 5: Stemming (fast, rough root-chopping)
stemmer = PorterStemmer()
stemmed = [stemmer.stem(t) for t in tokens]
print("Stremmed:", stemmed)
```

```python
# Step 6: Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(t) for t in tokens]
print("Lemmatized:", lemmatized)
```
